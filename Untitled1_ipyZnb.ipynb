```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Year": [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    "SubTotal": [1858769, 2110519, 2366626, 2602630, 2820649, 3015244, 3310211, 3440693, 3586805, 3832188, 4135292, 4475143, 4596690, 4961143, 5388022, 5864373, 6424917, 6480521],
    "Old_Age": [1517649, 1731560, 1949867, 2149168, 2330128, 2489614, 2748455, 2840660, 2947422, 3151349, 3412350, 3706516, 3778824, 4090497, 4468126, 4894452, 5396729, 5445159],
    "Disability": [61762, 67091, 72166, 74535, 76280, 75895, 75934, 75041, 75387, 75688, 75497, 75486, 75734, 77872, 78079, 77726, 76718, 70979],
    "BereavedFamily": [279358, 311868, 344593, 378927, 414241, 449735, 485822, 524992, 563996, 605151, 647445, 693141, 742132, 792774, 841817, 892195, 951470, 964383],
}

df = pd.DataFrame(data)


df["Old_Age_Percentage"] = df["Old_Age"] / df["SubTotal"] * 100
df["Disability_Percentage"] = df["Disability"] / df["SubTotal"] * 100
df["BereavedFamily_Percentage"] = df["BereavedFamily"] / df["SubTotal"] * 100


plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Old_Age_Percentage"], marker='o', label='Old Age Percentage')
plt.plot(df["Year"], df["Disability_Percentage"], marker='o', label='Disability Percentage')
plt.plot(df["Year"], df["BereavedFamily_Percentage"], marker='o', label='Bereaved Family Percentage')

plt.title('Yearly Percentage of SubTotal Categories')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'EconomicSentiment.xlsx'
df = pd.read_excel(file_path, index_col=0)


years = df.columns
economic_sentiment_original = df.loc['economic_sentiment(OriginalSeries)']
economic_sentiment_cyclical = df.loc['Economic_sentiment_index(CyclicalFluctuation)']


plt.figure(figsize=(10, 6))

plt.plot(years, economic_sentiment_original, marker='o', label='Original Series')
plt.plot(years, economic_sentiment_cyclical, marker='o', label='Cyclical Fluctuation')

plt.title('Economic Sentiment Over the Years')
plt.xlabel('Years')
plt.ylabel('Sentiment Value')
plt.legend()
plt.grid(True)

plt.show()
```

```python
df_defense = read_excel('NumberOfEnlistees.xlsx')
df_defense
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'NumberOfEnlistees.xlsx'
df = pd.read_excel(file_path)

years = df['Year']
enlistees = df['Number of enlistees']

min_enlistees = min(enlistees)

explode = [0.1 if enlistee == min_enlistees else 0 for enlistee in enlistees]

plt.figure(figsize=(8, 8))
plt.pie(enlistees, labels=years, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Number of Enlistees by Year')

plt.show()
```
